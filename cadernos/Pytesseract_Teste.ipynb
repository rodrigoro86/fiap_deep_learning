```python
import cv2
import numpy as np
import pytesseract
import matplotlib.pyplot as plt
```

![](../arq/CNH_Aberta/00000001_in.jpg)

```python
def print_image(image):
    plt.figure(figsize=(10, 5))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Converter BGR para RGB
    plt.axis('off')  # Remover eixos
```

```python
imagem = cv2.imread("../arq/CNH_Aberta/00003638_in.jpg")
```

```python
# Criar uma máscara preta do mesmo tamanho da imagem
mascara = np.zeros_like(imagem, dtype=np.uint8)
areas = [
    (80, 100, 400, 150),  # Nome (x1, y1, x2, y2)
    (320, 190, 500, 245),  # CPF
]
# Desenhar retângulos brancos nas áreas a serem lidas
for (x1, y1, x2, y2) in areas:
    cv2.rectangle(mascara, (x1, y1), (x2, y2), (255, 255, 255), -1)

# Aplicar a máscara à imagem original
imagem_mascarada = cv2.bitwise_and(imagem, mascara)

# Converter a imagem mascarada para escala de cinza
imagem = cv2.cvtColor(imagem_mascarada, cv2.COLOR_BGR2GRAY)
print_image(imagem)
```

```python
# Aplicar OCR na imagem mascarada
texto = pytesseract.image_to_string(imagem, lang='por')
```

```python
# Exibir resultado
print("Texto extraído:\n", texto)
```

```python
# Criar leitor do EasyOCR
reader = easyocr.Reader(['pt'])

# Fazer OCR
result = reader.readtext(imagem)
```

```python
for (bbox, text, prob) in result:
    text = text.strip().upper()  # Normalizar texto para facilitar buscas
    print(text)
```
